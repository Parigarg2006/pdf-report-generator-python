```python
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, Image
from reportlab.lib.units import inch, cm
from reportlab.pdfgen import canvas
import matplotlib.pyplot as plt
import numpy as np
import io

def create_chart():
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))
    months = ['Jan','Feb','Mar','Apr','May','Jun']
    sales = [45000, 52000, 48000, 61000, 55000, 67000]
    ax1.bar(months, sales, color='steelblue')
    ax1.set_title('Monthly Sales 2024')
    ax1.set_ylabel('в‚№ Revenue')

    categories = ['Electronics', 'Clothing', 'Food', 'Books', 'Sports']
    sizes = [35, 25, 20, 12, 8]
    ax2.pie(sizes, labels=categories, autopct='%1.0f%%', startangle=90)
    ax2.set_title('Sales by Category')

    buf = io.BytesIO()
    plt.tight_layout()
    plt.savefig(buf, format='png', dpi=100)
    buf.seek(0)
    plt.close()
    return buf

def generate_report(filename="sales_report.pdf"):
    doc = SimpleDocTemplate(filename, pagesize=A4,
                            rightMargin=2*cm, leftMargin=2*cm,
                            topMargin=2*cm, bottomMargin=2*cm)
    styles = getSampleStyleSheet()
    story = []

    # Define header and footer function
    def _header_footer(canvas_obj, doc_obj):
        canvas_obj.saveState()
        # Header
        canvas_obj.setFont('Helvetica-Bold', 10)
        canvas_obj.drawString(cm, A4[1] - cm, "Sales Performance Report")

        # Footer
        canvas_obj.drawString(cm, cm, "Page %d" % doc_obj.page)
        canvas_obj.restoreState()

    # Title
    title_style = ParagraphStyle('title', fontSize=20, fontName='Helvetica-Bold',
                                  spaceAfter=12, textColor=colors.HexColor('#1a73e8'))
    story.append(Paragraph("Sales Performance Report вЂ” Q1 2024", title_style))
    story.append(Paragraph("Generated by Python ReportLab", styles['Normal']))
    story.append(Spacer(1, 0.3*inch))

    # Chart
    chart_buf = create_chart()
    story.append(Image(chart_buf, width=6*inch, height=2.5*inch))
    story.append(Spacer(1, 0.2*inch))

    # Table
    data = [['Product', 'Units Sold', 'Revenue', 'Growth %'],
            ['Samsung TV', '1,245', 'в‚№3,98,400', '+12%'],
            ['iPhone 15', '892', 'в‚№7,13,600', '+8%'],
            ['Nike Shoes', '2,341', 'в‚№1,17,050', '+22%'],
            ['Books Bundle', '4,521', 'в‚№90,420', '+5%'],
            ['TOTAL', '9,999', 'в‚№13,19,470', '+13%']]

    table = Table(data, colWidths=[2.5*inch, 1.5*inch, 1.5*inch, 1.2*inch])
    table.setStyle(TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.HexColor('#1a73e8')),
        ('TEXTCOLOR', (0,0), (-1,0), colors.white),
        ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
        ('FONTSIZE', (0,0), (-1,-1), 10),
        ('ALIGN', (1,0), (-1,-1), 'CENTER'),
        ('ROWBACKGROUNDS', (0,1), (-1,-2), [colors.white, colors.HexColor('#f8f9fa')]),
        ('BACKGROUND', (0,-1), (-1,-1), colors.HexColor('#e8f0fe')),
        ('FONTNAME', (0,-1), (-1,-1), 'Helvetica-Bold'),
        ('GRID', (0,0), (-1,-1), 0.5, colors.HexColor('#dadce0')),
        ('ROWHEIGHT', (0,0), (-1,-1), 20),
    ]))
    story.append(table)
    doc.build(story, onFirstPage=_header_footer, onLaterPages=_header_footer)
    print(f"вњ… PDF Report saved: {filename}")

generate_report("sales_report.pdf")
```
